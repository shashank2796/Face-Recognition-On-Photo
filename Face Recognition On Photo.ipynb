```python
import cv2
import sys
import matplotlib.pyplot as plt
```

```python
# Get user supplied values
imagePath = sys.argv[1]
cascPath = r"E:\haarcascade_frontalface_default.xml"
```

```python
# Create the haar cascade
# cascade is just an XML file that contains the data to detect faces.
faceCascade = cv2.CascadeClassifier(cascPath)
```

```python
# Read the image and convert it to grayscale
image = cv2.imread(r'C:\Users\baps.DESKTOP-U41EIL5\Desktop\index.jfif')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image)
```

```python
# Detect faces in the image
# detectMultiScale function is a general function that detects objects
# scaleFactor - Some faces may be closer to the camera, they would appear bigger than the faces in the back
# minNeighbors defines how many objects are detected near the current one
# minSize - gives the size of each window.

faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    flags = cv2.CASCADE_SCALE_IMAGE 
)
```

```python
print("Found {0} faces!".format(len(faces)))
```

```python
# Draw a rectangle around the faces
# function returns 4 values:x,y location of the rectangle, and w,h width and height.
# We use these values to draw a rectangle using the built-in rectangle() function.
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
cv2.imshow("Faces found", image)
cv2.waitKey(0)
```
